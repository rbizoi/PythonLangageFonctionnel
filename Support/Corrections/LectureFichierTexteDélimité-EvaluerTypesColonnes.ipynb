# Décorateur temps d'exécution

```python
import time

def tempsExecution(func):
    def chronometrerExécution(*_args, **_kwargs): 
        t0 = time.perf_counter()
        result = func(*_args, **_kwargs) 
        elapsed = time.perf_counter() - t0
        name = func.__name__
        args = ', '.join(repr(arg) for arg in _args) \
               if len(_args) == 1 \
               else ', '.join(arg+'='+repr(_kwargs[arg]) for arg in _kwargs)
        print(f'exécution {elapsed:0.8f}s\nfonction {name}({args})')
        return result
    return chronometrerExécution
```

# Lecture fichier et création du dictionnaire

```python
@tempsExecution
def lectureFichier(nom,caractere=';', listNa = ['mq','']):
    
    if os.path.isfile(nom) :
        with open(nom,'r') as fichier :
            colonnes=fichier.readline().replace('\n','').split(caractere)
            while '' in colonnes : colonnes.pop()
            dictionnaire = { cle:list() for cle in colonnes}
            ligne =fichier.readline()
            while ligne :
                ligne = ligne[:-1].split(caractere)[:len(colonnes)] if ligne[-1] == '\n' else ligne.split(caractere)[:len(colonnes)]
                _ = [ dictionnaire[cle].append(None if ligne[i] in listNa else ligne[i]) 
                      for i,cle in enumerate(dictionnaire) ]
                ligne =fichier.readline()
        return dictionnaire  
    else : 
        return {}  
```

```python
def initTypeColones(dictionnaire)-> (dict):
    dictionnaire = {colonne:
                     [int(i) for i in dictionnaire[colonne]]
                     if sum([str(i).isdecimal() for i in dictionnaire[colonne]]) == len(dictionnaire[colonne]) 
                     else [float(i) if i is not None else None for i in dictionnaire[colonne]] 
                            if sum([re.sub('[+-\.]','',str(i), count=1).isdecimal() or (i is None) for i in dictionnaire[colonne]]) 
                                   == len(dictionnaire[colonne]) 
                            else dictionnaire[colonne]
                 for colonne in dictionnaire }
    return dictionnaire
```

```python
dictionnaire = initTypeColones(dictionnaire)
```
